```python
#correlation matrix
corr_matrix=data.corr()
plt.figure(figsize=(14,14))  #14in by 14in
sns.heatmap(corr_matrix, annot=True, fmt='.0%')
```

```python
#Transform non-numeric columns into numerical columns
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
for column in data.columns:
        if data[column].dtype == 'int64':
            continue
        data[column] = LE.fit_transform(data[column])
```

```python
data.head(10)
```

```python
Y=data.iloc[:,0]
X=data.iloc[:,1:]
#Splitting data into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,train_size=0.2,random_state=0)
```

```python
# Train a few models and compare accuracies
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier

models=[RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0), LogisticRegression(),XGBClassifier(random_state=0)]
```

```python
accuracy_train=[]
accuracy_test=[]
from sklearn.metrics import accuracy_score
for model in models:
    Model=model
    Model.fit(X_train,y_train)
    accuracy_train.append(accuracy_score(y_train,Model.predict(X_train)))
    accuracy_test.append(accuracy_score(y_test,Model.predict(X_test)))
print(accuracy_test)
print(accuracy_train)
```

```python
#Random Forest performs better,tuning the model
Model=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
print('Parameters currently in use:')
print(Model.get_params())
```

```python
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,'max_features': max_features,'max_depth': max_depth,'min_samples_split': min_samples_split,'min_samples_leaf': min_samples_leaf,'bootstrap': bootstrap}
print(random_grid)
```

```python
rf = RandomForestClassifier()
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
rf_random.fit(X_train, y_train)
```

```python
rf_random.best_params_
```

```python
final_model=RandomForestClassifier(n_estimators= 600,
 min_samples_split= 5,
 min_samples_leaf= 1,
 max_features= 'sqrt',
 max_depth= 60,
 bootstrap= False)
final_model.fit(X_train,y_train)
print('Test accuracy after tuning: ',accuracy_score(y_test,final_model.predict(X_test)))
print('Increase in performance:', (accuracy_score(y_test,final_model.predict(X_test))-accuracy_test[0])/accuracy_test[0]*100)
```

```python
# Return the feature importances (the higher, the more important the feature).
importances = pd.DataFrame({'feature':data.iloc[:, 1:data.shape[1]].columns,'importance':np.round(final_model.feature_importances_,3)}) #Note: The target column is at position 0
importances = importances.sort_values('importance',ascending=False)
importances
```

```python
plt.figure(figsize=(20,8))
plt.bar(list(importances['feature']),list(importances['importance']))
plt.xticks(rotation=90)
plt.show()
```
